# Benchmark functions

The implemented functions are:
- Ackley: $f\left(\mathbf{x}\right) = -a\exp\left(-b\sqrt{\frac{1}{d}\sum\limits_{i=0}^{d}x_i^2}\right) - \exp\left(\frac{1}{d}\sum\limits_{i=0}^{d}\cos\left(c x_i\right)\right) + a +\exp\left(1\right)$
- Cigar: $10^{-6}x_0^2 + \sum\limits_{i=1}^{d}x_i^2$
- Diffpowers: $\sum\limits_{i=0}^{d}|x_i|^{\left(2 + \frac{10i}{d-1}\right)}$
- Discus: $x_0^2 + \sum\limits_{i=1}^{d} 10^{-6}x_i^2$
- Ellipsoid: $\sum\limits_{i=0}^{d}10^{\frac{-6i}{d-1}} x_i^2$
- Rosenbrock: $\sum\limits_{i=0}^{d-1}\left(100\left(x_{i+1} - x_i^2\right)^2 + \left(1-x_i^{\phantom{2}}\right)\right)$
- Sphere: $||\mathbf{x}||^{2}$
